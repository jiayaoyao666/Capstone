```python
#获取新闻数据的接口
import pandas as pd
import numpy as np
import tushare as ts
```

```python
from datetime import datetime, timedelta
start_date = datetime(2018, 10, 11)
end_date = datetime(2024, 3, 20)
dates_list = []
current_date = start_date
while current_date <= end_date:
    dates_list.append(current_date)
    current_date += timedelta(days=1)
```

```python
# 打印列表中的日期
for date in dates_list:
    print(date.strftime('%Y-%m-%d'))
```

```python
standard_dates_list=[date.strftime('%Y-%m-%d') for date in dates_list]
standard_dates_list[:5]
```

```python
#test
pro = ts.pro_api("647b0813cd4fae53c8d2bac0dc72b98abe5125b85dd807ca5f9b2d4e")
df = pro.news(src='sina', start_date='2018-10-10', end_date='2018-10-11')
df
```

```python
#build dataframe and merge
news_dataframe=pd.DataFrame()
for i in range(len(standard_dates_list)-1):
    df = pro.news(src='sina', start_date=standard_dates_list[i], end_date=standard_dates_list[i+1])
    print(standard_dates_list[i])
    news_dataframe = pd.concat([news_dataframe, df], ignore_index=True)
```

```python
news_dataframe.to_csv(r"./新闻快讯.csv",index=False,encoding='utf_8_sig')
```
