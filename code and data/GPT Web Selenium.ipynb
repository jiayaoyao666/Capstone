```python
import pandas as pd
import numpy as np
```

```python
news=pd.read_csv(r"./新闻快讯_新浪.csv")
news
```

```python
news['new_column'] = news.apply(lambda row: f"第{row.name+1}条:{row['content']}", axis=1)
news
```

```python
#提示词工程：你是一个中立的金融学者，可以判断财经新闻的种类和信息。现在我将输入新闻标题，请你输出新闻特征
# 模拟在网页上输入
search_box = driver.find_element('class name', 'GrowingTextArea_textArea__ZWQbP')
search_box.clear()  # 清除搜索框中的默认文本
search_box.send_keys('你是一个中立的金融学者，可以判断财经新闻的种类。微观新闻是指新闻是否与行业或者某家公司有关,或者宏观新闻是指新闻与政府政策、财政预算、税收政策、商品价格、产业概况、股票指数、汇率、利率、期货价格、贵金属、GDP、通胀、就业等有关。现在我将输入多条有序号的新闻，请你只输出"宏观"或"微观"，每条回答用"|"隔开，不要输出其他信息。我会持续输入，你也要持续输出')
# 提交搜索
search_box.send_keys(Keys.RETURN)
```

```python
sample_size=10
sample=news.iloc[:sample_size,:]
total_content=""
for i in range(sample_size):#len(news
    content=news.iloc[i,3]
    total_content+=content
search_box = driver.find_element('class name', 'GrowingTextArea_textArea__ZWQbP')
search_box.clear()  # 清除搜索框中的默认文本
search_box.send_keys(total_content)
# 提交搜索
search_box.send_keys(Keys.RETURN)
```

```python
import time
```

```python
#循环提取信息
#选前100条数据
sample=news.iloc[:100,:]
batch_size=5
hint="你是一个专业的金融从业者，可以判断财经新闻的种类。微观新闻是指新闻与行业或者某家公司有关,宏观新闻是指新闻与政府政策、商品价格、股票指数、汇率、利率、贵金属、GDP、通胀、就业等有关。现在我将输入多条有序号的新闻，请你只输出'宏观'或'微观'，每条回答用'|'隔开，不要输出其他信息。下面我将输入新闻内容，每一条新闻前都有数字编号，有的新闻句子较多，但也视为一条新闻。"
for episode in range(0,20):
    total_content=""
    for i in range(batch_size):
        content=sample.iloc[batch_size*episode+i,-1]
        total_content+=content
    print(episode)
    #输入新闻
    search_box = driver.find_element('class name', 'GrowingTextArea_textArea__ZWQbP')
    search_box.clear()  # 清除搜索框中的默认文本
    search_box.send_keys(hint+total_content)
    search_box.send_keys(Keys.RETURN)
    #生成(0,1)的随机数
    sleep_time=np.random.rand()+5
    time.sleep(sleep_time)
    print(episode)
```

```python
# 指定Chrome的远程调试地址
chrome_options = Options()
chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9223")
# 指定ChromeDriver的路径
chrome_driver_path = "D:\\software\\anaconda\\chromedriver-win64\\chromedriver.exe"
service = Service(executable_path=chrome_driver_path)
# 创建WebDriver对象
driver = webdriver.Chrome(service=service, options=chrome_options)
# 现在可以控制已打开的浏览器了
# 访问网页
driver.get('https://poe.com')#当前对话的网址
```

```python
# 使用 execute_script 方法滚动到元素可见的最上端
element = driver.find_element(By.CLASS_NAME, "Markdown_markdownContainer__Tz3HQ")
driver.execute_script("arguments[0].scrollIntoView(true);", element)
#开始爬取
# 获取所有具有指定类名的元素
elements = driver.find_elements(By.CLASS_NAME, "Markdown_markdownContainer__Tz3HQ")
# 遍历所有找到的元素，并获取它们的文本内容
for element in elements:
    text = element.text
    print(text)
```
